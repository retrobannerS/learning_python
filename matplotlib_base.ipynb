### Построение обычного графика.

Для начала линейный график. Будем строить его по точкам, для этого нужна команда `plt.plot()`.

```python
import numpy as np
x = np.linspace(1, 2, 2)
y = x

plt.title("Линейная зависимость y = x")
plt.xlabel("x - ось абсцисс")
plt.ylabel("y - ось ординат")
plt.grid()
plt.plot(x, y)
```

Изменим цвет линии, добавив третий аргумент. `r <=> red`, `-- <=> пунктиром`.

```python
plt.plot(x, y, "r--")
```

Несколько графиков на одном поле.

```python
x = np.linspace(0, 10, 50)
y1 = x
y2 = x * x

plt.title("y1 = x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()
plt.plot(x, y1, "g-", x, y2, "b-")
```

Несколько полей. `figure` задает параметры для отображения графиков. `subplot` задает местоположение поля с графиком. Первый аргумент - количество строк, второй - количество столбцов, третий - индекс (номер поля сверху вниз, слева направо).

```python
plt.figure(figsize=(9, 9))

plt.subplot(2, 1, 1)
plt.plot(x, y1)               # построение графика
plt.title("Зависимости: y1 = x, y2 = x^2") # заголовок
plt.ylabel("y1", fontsize=14) # ось ординат
plt.grid()
plt.subplot(2, 1, 2)
plt.plot(x, y2)               # построение графика
plt.xlabel("x", fontsize=14)  # ось абсцисс
plt.ylabel("y2", fontsize=14) # ось ординат
plt.grid()                    # включение отображение сетки
```

Мы дополнительно задали размер шрифта в подписях к осям.

### Построение гистограммы

Здесь будет работать функция `plt.bar()`.

```python
choices = ["Да", "Нет", "Не знаю"]
counts = [1, 4, 6]
plt.title("Итоги голосования")
plt.ylabel("Количество")
plt.ylabel("Варианты ответа")
plt.bar(choices, counts)
```

Приведу основные составляющие изображения в `matplotlib`.

1. Фигура (figure)
2. Заголовок (title)
3. Подписи осей x и y (x/ylabel)
4. Легенда (legend)
5. Основная сетка (major grid) и дополнительная сетка (minor grid) 
6. Основные тики (major ticks) и дополнительные тики (minor ticks) - промежуток между значениями на осях.

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
x = np.linspace(0, 10, 10)
y1 = 4*x
y2 = x * x
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title("Графики зависимостей: y1=4*x, y2=x^2", fontsize=16)
ax.set_xlabel("x", fontsize=14)        
ax.set_ylabel("y1, y2", fontsize=14)
ax.grid(which="major", linewidth=1.2)
ax.grid(which="minor", linestyle="--", color="gray", linewidth=0.5)
ax.scatter(x, y1, c="red", label="y1 = 4*x") # Точечный график
ax.plot(x, y2, label="y2 = x^2") # сплошной график
ax.legend()
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)
plt.show()
```
